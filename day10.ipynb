 # input output creation

```python
ip=advr.drop('Sales',axis=1)
op=advr.Sales
```

# train test split: splitting the data for training and testing the machine

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(ip,op,test_size=0.2)
```

# standard scalar: standardize the data or scaling the data to a certain range

formula :


  z= x - M(miu)/ sigma

```python
from sklearn.preprocessing import StandardScaler 

```

```python
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
x_train
```

```python
x_test
```

# linear regression :

it finds the linear relationship between dependent and independent variables

it is used to create a best fit line which will consider all the data points or we can say maximum points should lie on the best fit line

dependent varible = output ( x axis )

independent variable= input ( y axis)

the formula used in linear regression is straight line equation i.e


y= mx+c

where ,

x= independent

y= dependent

m= slope

c= y-intercept

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
y_pred
```

```python
y_test
```

# mean squared error



```python
from sklearn.metrics import mean_squared_error,r2_score

mse=mean_squared_error(y_pred,y_test)
r2=r2_score(y_pred,y_test)
print("mean squared error :",mse)
print("accuracy",r2)
```

```python
#lm plot: linear model plot 
df=pd.DataFrame({'y_test(original_answer)':list(y_test),
                 'y_pred(machine_generate answer)':list(y_pred)})
df
```

```python
import seaborn as sns
```

```python
sns.lmplot(x='y_test(original_answer)',y='y_pred(machine_generate answer)',data=df)
```
