# filling nan values with the type 2 column

```python
pkmn.fillna({'Type 2' :pkmn['Type 2'].mode()[0]},inplace=True)
```

```python
pkmn
```

```python
pkmn.info()
```

```python
for i in pkmn.columns:
    print(i,":",pkmn[i].unique())
```

```python
pkmn.columns
```

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()
pkmn.Name=le.fit_transform(pkmn.Name)
```

```python
pkmn["Type 1"]=le.fit_transform(pkmn["Type 1"])
```

```python
pkmn["Type 2"]=le.fit_transform(pkmn["Type 2"])
```

```python
pkmn["Legendary"]=le.fit_transform(pkmn["Legendary"])
```

```python
pkmn
```

```python
pkmn['Legendary'].unique()
```

```python
pkmn['Legendary'].value_counts()
```

```python
x=pkmn[['Name','Type 1', 'Type 2', 'Combat Power', 'HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed', 'Generation']]
y=pkmn['Legendary']
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train
```

```python
x_test
```

```python
x_test
```

# standard scalling

```python
from sklearn.preprocessing import StandardScaler 
```

```python
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

# logistic regression

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
y_pred
```

```python
y_test
```

```python
df=pd.DataFrame({'y_test(original_answer)':list(y_test),
                 'y_pred(machine_generate answer)':list(y_pred)
                })
df
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
```

```python
acc=accuracy_score(y_pred,y_test)
pre=precision_score(y_pred,y_test,zero_division=0)
rec=recall_score(y_pred,y_test,zero_division=0)
f1=f1_score(y_pred,y_test,zero_division=0)
```

```python
print("ACCURACY",acc)
print('PRECISION',pre)
print('recall',rec)
print('f1_score',f1)
```

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
```

```python
cm=confusion_matrix(y_pred,y_test)
cm
```

```python
cmd=ConfusionMatrixDisplay(cm)
cmd.plot()
```
