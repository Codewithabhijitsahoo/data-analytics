# matplotlib library

```python
#import matplotlib packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# line plot


```python
x=[1,2,3,4,5,6]
y=[12,13,14,15,23,16]
plt.plot(x,y)
plt.show()
```

```python
x=[1,4,5,6,7,8,9]
y=[22,24,55,66,88,99,100]
plt.plot(x,y)
plt.show()
```

# hist plot

```python
age=[11,22,55,66,34,22,55,6,22,15,4,57,85,88]
plt.hist(age,bins=4,width=8)
plt.title("histogram")
plt.grid()
plt.xlabel("x-axis")
plt.ylabel("y-axis")
```

# bar graph

```python

x=[1,2,3,4]
y=[5,6,7,8]
plt.bar(x,y)

plt.show()
```

```python
x=[1,2,3,4]
y=[5,6,7,8]
plt.bar(x,y,color="r")

plt.show()
```

```python
x=[1,2,3,4]
y=[5,6,7,8]
color=["r","b","y"]
plt.bar(x,y,color=color)

plt.show()
```

# pie chart

```python
work=["phone","eating","sleeping",'working','studying']
time=[6,8,2,5,4]
t=[0,0.1,0,0,0]
plt.title("pie_chart")
plt.pie(time,labels=work,explode=t,autopct='%1.1f%%')
plt.show()
```
