# pandas library:

Pandas is a powerful and widely used Python library for data manipulation and analysis, especially when working with structured data like tables or spreadsheets. Think of it as your go-to toolkit for cleaning, transforming, and exploring data efficiently.
At its core, pandas revolves around two main data structures:

- Series: A one-dimensional labeled array (like a single column).

- DataFrame: A two-dimensional labeled data structure (like an Excel sheet with rows and columns).or contain multiple columns



```python
import pandas as pd
```

```python
#isnull():
import numpy as np
dict={
    "names":["abhi","rahul","aditya","somya"],
    "roll_no":[34,33,23,np.nan],
    "marks":[66,55,77,78]
}
data=pd.DataFrame(dict)
data.isnull()
```

```python
#fillna():to put value inplace of the null values
import pandas as pd
dict={
    "names":["abhi","rahul","aditya","somya"],
    "roll_no":[34,33,23,np.nan],
    "marks":[66,55,77,78]
}
data=pd.DataFrame(dict)

x=data.fillna(3)
print(x)
```

```python
#rename(): to rename the column name 
data=data.rename(columns={"names":"user_name"})
data
```
