# seaborn

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
iris=pd.read_csv(r"D:\iris.data",header=None)
iris
```

```python
iris.columns=['sl','sw','pl','pw','flower']
```

```python
iris.columns
```

```python
iris.head()
```

```python
#scatter plot
sns.scatterplot(x=iris.sw,y=iris.sl)
plt.show()
```

```python
sns.scatterplot(x=iris.pl,y=iris.pw)
plt.show()
```

```python
#dist plot(dis plot + kde plot ): use to display the distribution of data points
sns.distplot(iris.sl)
plt.show()
```

```python
#dis plot(): use to display teh distribution of  a single variable using histogram
sns.displot(iris.sl)
plt.show()
```

```python
#box plot(): use to see statistical information 
sns.boxplot(x=iris.flower,y=iris.sw,hue=iris.flower)
plt.show()
```

```python
sns.boxplot(x=iris.flower,y=iris.sw,palette='cool')
plt.show()
```

```python
#boxen plot

sns.boxenplot(x=iris.flower,y=iris.sw,hue=iris.flower)
plt.show()
```

```python
# bar plot: use display yhe mean of a variable along with the categorical data
sns.barplot(x=iris.flower,y=iris.sw,hue=iris.flower)
plt.show()
```

```python
#count plot(): show the count of each categories
sns.countplot(x=iris.flower,hue=iris.flower)
plt.show()
```

```python
#pair plot: show relationship  between contineous column with contineous column
sns.pairplot(data=iris)
plt.show()
```

```python
sns.pairplot(data=iris,hue='flower')
plt.show()
```
